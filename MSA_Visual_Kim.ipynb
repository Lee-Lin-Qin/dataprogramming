```python
import pandas as pd
# import numpy as np                  # We don't need numpy in this visualization.
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_rows=None
pd.options.display.max_columns=None
```

```python
train_df =  pd.read_csv('train.csv',sep=',',header=0)          ## train.csv
train_df.head(3)
```

```python
scatterplot1 = sns.relplot(
    x = "Age",                      
    y = "Previously_Insured",
    data = train_df,            # Using data from 'train_df'
    hue = "Response",           # Show different color by 'Response'        
)
plt.show(scatterplot1)
plt.close()
```

```python
scatterplot1.savefig('Age_PreInsured.png')         # To save a result in png file
```

```python
scatterplot4 = sns.relplot(
    x = "Age",
    y = "Annual_Premium",
    data = train_df.sample(3800, random_state=123),  # Use only 3800 samples from train_df
    hue = "Response",
    #size = "Previously_Insured",  # When I add this, the result looks too busy.
    #sizes = (50,250),           # (minimum, maximum)
    style = "Gender",           # Use different symbols by Gender
    height = 8,                 # height of graph
    aspect = 1.5,               # ratio of graph
)

plt.show(scatterplot4)
plt.close()
```

```python
scatterplot4.savefig('Age_Annual_Premium.png')
```

```python
scatterplot5 = sns.relplot(
    x = "Gender",
    y = "Age",
    data = train_df,
    hue = "Response",            
)
plt.show(scatterplot5)
plt.close()
```

```python
scatterplot5.savefig('Gender_Age.png')
```
